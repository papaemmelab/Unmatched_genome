# - Annotation: Stops

```python
stop_list='/work/isabl/home/gutierj2/uk_all/notebooks/SNVs/TSGs/*_2.output.annot.stop.bcftools.tsv'
stop_list=!ls {stop_list}
print(len(stop_list))
stop_list
```

```python
stop_total=pd.DataFrame()
for n in stop_list:
    sample=n.split('/')[9].split('_2')[0]
    stop_ann=pd.read_csv(n,sep='\t',comment='#')
    stop_ann=stop_ann[['CHR','START','REF','ALT','QUAL','VAG_GENE','VAG_cDNA_CHANGE','VAG_PROTEIN_CHANGE','VEP_Consequence','VEP_IMPACT']]
    stop_ann=stop_ann.drop_duplicates(subset=None, keep='first', inplace=False)
    stop_ann['VAG_PROTEIN_CHANGE']=stop_ann['VAG_PROTEIN_CHANGE'].str.replace('p.', '', regex=False)
    stop_ann['Sample']=sample
    stop_total=stop_total.append(stop_ann)
```

```python
stop_total.sort_values(by=['Sample'])
```

```python
keys=[]
for index,row in stop_total.iterrows():
    key=str(row['CHR'])+'_'+str(row['START'])+'_'+row['Sample']
    keys.append(key)
stop_total['keys']=keys
```

```python
samples_noDetect=list(stop_total['Sample'].unique())
SV_support_stops=list()
SV_no_support_stops=list()

for n in samples_noDetect:
    #isabl
    list_annotSVs=!isabl get-outdirs -fi application__name ANNOT_SNVS -fi targets__system_id {n}
    path=list_annotSVs[1]+'/drivers/*.snvs.output.annot.coding.tsv.gz'
    print(n)

    svs_pass=!ls {path}
    
    #iterate df
    df=stop_total[stop_total['Sample']==n]
    df=df.drop_duplicates(subset=['keys'])
    #print(df.shape)
    #print('- ',n,':','\n',df[['chr_seg','start_seg','end_seg','CNV','Sample','Fraction']],'\n')
    chrom_list=list()
    
    #SV_support=list()
    for index,row in df.iterrows():
        gene=row['VAG_GENE']
        alteration=row['VAG_PROTEIN_CHANGE']

        
        #Awk command + print
        awk_output=!zless {svs_pass[0]} | grep {gene} | grep {alteration}
        awk_output=list(set(awk_output))
        #print(awk_output)
        #if (not chrom in chrom_list):
            #for n2 in awk_output:
                #print(n2)
            #print('\n')
            #chrom_list.append(chrom)
        
        #SV support
        
        if(len(awk_output)==0):
            SV_no_support_stops.append(row['keys'])
        else:            
            SV_support_stops.append(row['keys'])

print('SVsupport_stops:')
SV_support_stops=list(set(SV_support_stops))
print(len(SV_support_stops))
#for n3 in sorted(SV_support):
    #print(n3)
print('No SVsupport_stops:')
SV_no_support_stops=list(set(SV_no_support_stops))
print(len(SV_no_support_stops))
#for n3 in sorted(SV_no_support):
    #print(n3)
```

```python
stop_total_matched=stop_total.copy()

match_support=[]
for index,row in stop_total_matched.iterrows():
    if(row['keys'] in SV_support_stops):
        match_support.append('Detected')
    else:
        match_support.append('Non Detected')
stop_total_matched['Matched']=match_support
```
